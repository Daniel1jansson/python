# In this mini-project I have done some data cleaning and exploration of a dataframe that consists of salary data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

plt.style.use("seaborn-whitegrid")
colors = ["#54436B", "#50CB93", "#ED8E7C", "#FFD523", '#00C1D4', '#F8485E']
sns.set_palette(sns.color_palette(colors))
```

```python
missing_values = [np.nan, 'nan', 'N/a']
df = pd.read_csv('Manager_salary.csv', na_values=missing_values)
```

# Checking null values

```python
df.isnull().sum()
```

```python
df
```

# In the Annual salary column, some values have commas. Commas are removed, so that all values are in the same format.

```python
df['Annual salary'] = df['Annual salary'].str.replace(',', '')
df['Annual salary'] = df['Annual salary'].astype(float)
```

```python
df
```

# In the 'Country' column it can be observed that United States is spelled differently, e.g in some instances it is spelled 'US'. 

```python
df['Country'] = df['Country'].str.replace('.', '')
df['Country'] = df['Country'].str.lower()
df['Country'] = df['Country'].str.strip()
countries = df['Country'].unique()
countries = sorted(countries)

countries = sorted(countries)

for y, x in enumerate(countries):
    print(y, x)
```

```python
# list of different ways United States is spelled
United_States = countries[161:164] + countries[169:197] + countries[205:220] + countries[145:146]

L = [3, 79, 149]
for y, x in enumerate(countries):
    if y in L:
        United_States.append(x)
United_States
```

```python
Countries = []
for country in df['Country']:
    if country in United_States:
        Countries.append('United States')
    else:
        Countries.append(country)
df['Country'] = Countries
```

```python
df
```

# Now every every instance of United States is spelled 'United States'.

# The dataset is narrowed down to only include instances of the United States

```python
df = df[df['Country'] == 'United States']
df
```

```python
# Reset index
df.reset_index(drop=True, inplace=True)
```
